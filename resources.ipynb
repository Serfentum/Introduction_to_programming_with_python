```python
def linear_search(xs, ele):
    for i, elem in enumerate(xs):
        if elem == ele:
#             print(elem)
            return i

def binary_search(xs, elem):
    length = len(xs)
    prev = length // 2
    index = prev

    for i in range(round(length ** 0.5) + 1):
#         print(f'{i}th step')
#         print(xs[index])
        if xs[index] == elem:
            return index
        elif xs[index] < elem:
            index += round(prev / 2)
        else:
            index -= round(prev / 2)
        prev = round(prev / 2)
```

```python
%%timeit
linear_search(xs, 4545)
```

```python
%%timeit
binary_search(xs, 4545)
```
